```python
import pandas as pd

pd.set_option('max_column', 50)
nba_df = pd.read_csv("../Data/nba.csv")
nba_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(nba_df.corr())
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(14, 6))
plt.title("HomePCT vs AwayPCT")
plt.scatter(nba_df['HomePCT'], nba_df['AwayPCT'], alpha=0.5)
plt.plot([0, 1], [0, 1], color='r')
plt.xlabel("HomePCT")
plt.ylabel("AwayPCT")
plt.show()
```

```python
better_at_away = nba_df[nba_df['HomePCT'] < nba_df['AwayPCT']]
better_at_away
```

```python
soccer_df = pd.read_csv("../Data/soccer_leagues.csv")
soccer_df.head()
```

```python
soccer_df.shape
```

```python
soccer_df['Sport'].value_counts()
```

```python
soccer_df['Unnamed: 15'].value_counts()
```

```python
soccer_df.drop(['Sport', 'Unnamed: 15'], axis=1, inplace=True)
soccer_df.head()
```

```python
soccer_df['HomePCT'] = soccer_df['HomeWins'] / (soccer_df['HomeWins'] + soccer_df['HomeDraw'] + soccer_df['HomeLoss'])
soccer_df.head()
```

```python
soccer_df['AwayPCT'] = soccer_df['AwayWins'] / (soccer_df['AwayWins'] + soccer_df['AwayDraw'] + soccer_df['AwayLoss'])
soccer_df.head()
```

```python
plt.figure(figsize=(14, 6))
plt.title("HomePCT vs AwayPCT")
plt.scatter(soccer_df['HomePCT'], soccer_df['AwayPCT'], alpha=0.5)
plt.plot([0, 1], [0, 1], color='r')
plt.xlabel("HomePCT")
plt.ylabel("AwayPCT")
plt.show()
```

```python
nba_df['Home_Factor'] = (nba_df['HomePCT'] - nba_df['AwayPCT']) / (nba_df['HomePCT'] + nba_df['AwayPCT'])
soccer_df['Home_Factor'] = (soccer_df['HomePCT'] - soccer_df['AwayPCT']) / (soccer_df['HomePCT'] + soccer_df['AwayPCT'])
nba_df.head()
```

```python
soccer_df.head()
```

```python
plt.figure(figsize=(14, 6))
sns.kdeplot(soccer_df['Home_Factor'], shade=True)
sns.kdeplot(nba_df['Home_Factor'], shade=True)
plt.plot([0, 0],[0, 3])
plt.xlabel('Home Factor')
plt.legend(['Soccer Leagues', 'NBA'])
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.title("Home Average vs Away Average in Goals")
plt.scatter(soccer_df['HomeRatio'], soccer_df['AwayGoalsDiff'], alpha=0.5)
plt.xlim([-40, 40])
plt.ylim([-40, 40])
plt.plot([-40, 40], [-40, 40], color='r')
plt.xlabel("Home Diff")
plt.ylabel("Away Diff")
plt.show()
```

```python
import numpy as np

fn = np.polyfit(nba_df['Year'], nba_df['Home_Factor'], 1)
fit_fn = np.poly1d(fn)

plt.figure(figsize=(14, 6))
plt.scatter(nba_df['Year'], nba_df['Home_Factor'], alpha = 0.8)
plt.plot(nba_df['Year'], fit_fn(nba_df['Year']), color='r')
plt.ylim([-1, 1])
plt.show()
```

```python
country_df = pd.read_csv("../Data/country_facts.csv")
country_df.head()
```

```python
plt.figure(figsize = (14, 14))
plt.subplot(2,1,1)
plt.title('Home Advantage vs FIFA Rank')
plt.plot(country_df['FIFA_Rank'], country_df['Home_Away_Contrast'], 'o')
fit_fn = np.poly1d(np.polyfit(country_df['FIFA_Rank'], country_df['Home_Away_Contrast'], 1)) 
plt.plot(country_df['FIFA_Rank'], fit_fn(country_df['FIFA_Rank']))

plt.subplot(2,1,2)
plt.title('Home Advantage vs Crowd Attendance')
plt.plot(country_df['Attendance'], country_df['Home_Away_Contrast'], 'o')
fit_fn = np.poly1d(np.polyfit(country_df[country_df['Attendance'] > 0]['Attendance'], 
                              country_df[country_df['Attendance'] > 0]['Home_Away_Contrast'], 1)) 
plt.plot(country_df[country_df['Attendance'] > 0]['Attendance'], 
         fit_fn(country_df[country_df['Attendance'] > 0]['Attendance']))
plt.xlabel('Average Attandance per game in the league')
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(country_df.corr())
plt.show()
```

```python
plt.figure(figsize = (14, 14))
plt.subplot(2, 1, 1)
plt.title('Home Advantage vs # Phones per Person')
plt.scatter(country_df['Phones'], country_df['Home_Away_Contrast'])
fit_fn = np.poly1d(np.polyfit(country_df[country_df['Phones'] > 0]['Phones'], 
                              country_df[country_df['Phones'] > 0]['Home_Away_Contrast'], 1)) 
plt.plot(country_df[country_df['Phones'] > 0]['Phones'], 
         fit_fn(country_df[country_df['Phones'] > 0]['Phones']), color='r')

plt.subplot(2, 1, 2)
plt.title('Home Advantage vs Literacy')
plt.scatter(country_df['Literacy'], country_df['Home_Away_Contrast'])
fit_fn = np.poly1d(np.polyfit(country_df[country_df['Literacy'] > 0]['Literacy'], 
                              country_df[country_df['Literacy'] > 0]['Home_Away_Contrast'], 1)) 
plt.plot(country_df[country_df['Literacy'] > 0]['Literacy'], 
         fit_fn(country_df[country_df['Literacy'] > 0]['Literacy']), color='r')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.title('Home Advantage vs GDP')
plt.scatter(country_df['GDP'], country_df['Home_Away_Contrast'])
fit_fn = np.poly1d(np.polyfit(country_df[country_df['GDP'] > 0]['GDP'], 
                              country_df[country_df['GDP'] > 0]['Home_Away_Contrast'], 1)) 
plt.plot(country_df[country_df['GDP'] > 0]['GDP'], 
         fit_fn(country_df[country_df['GDP'] > 0]['GDP']), color='r')
plt.show()
```
