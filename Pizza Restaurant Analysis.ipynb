```python
df.drop(['country', 'menuPageURL', 'menus.currency', 'priceRangeCurrency'], axis=1, inplace=True)
df.head(2)
```

```python
df.drop(['keys', 'menus.dateSeen', 'priceRangeMin', 'priceRangeMax'], axis=1, inplace=True)
df.head(2)
```

```python
pizza_name = df.drop_duplicates(subset=['id', 'menus.name'])
pizza_name = pizza_name['menus.name'].value_counts()
pizza_name = pizza_name.to_frame()
pizza_name.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
plt.figure(figsize=(14, 3))
plt.title("Top 20 Pizza Name", fontsize=20)
plt.plot(pizza_name.head(20), linestyle='None', marker='o', markersize=10)
plt.xticks(rotation=90)
plt.show()
```

```python
province = df.drop_duplicates(subset=['id', 'province'])
province = province['province'].value_counts().to_frame()
province.head()
```

```python
plt.figure(figsize=(16, 6))
plt.title("Top 20 Pizza Locations", fontsize=20)
plt.plot(province.head(20), linestyle='None', marker='o', markersize=10, color='g')
plt.show()
```

```python
df[df['menus.amountMax'] == 0.0]
```

```python
df.drop(index=1869, inplace=True)
df[df['menus.amountMax'] == 0.0]
```

```python
amounts = df.drop_duplicates(subset=['id', 'menus.name'])
amounts = amounts.sort_values(by='menus.amountMax', ascending=False)
amounts.head(10)
```

```python
amounts = amounts.groupby('province')['menus.amountMax'].mean().nlargest(10).to_frame()
amounts.head()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8, 4))
plt.title("Top expensive provinces", fontsize=20)
plt.plot(amounts['menus.amountMax'].head(7), linestyle='', marker='o', markersize=10, color='darkred')
plt.xticks(rotation=90)
plt.show()
```

```python
amounts = amounts.groupby('province')['menus.amountMax'].mean().nsmallest(10).to_frame()
amounts.head()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8, 4))
plt.title("Top cheap provinces", fontsize=20)
plt.plot(amounts['menus.amountMax'].head(7), linestyle='', marker='o', markersize=10, color='darkred')
plt.xticks(rotation=90)
plt.show()
```

```python
province_by_count = df['province'].value_counts()
province_by_count = province_by_count.nlargest(20).to_frame()
province_by_count.head()
```

```python
plt.figure(figsize=(14, 6))
plt.title("Top pizza locations", fontsize=20)
plt.plot(province_by_count, linestyle='', marker='o', markersize=10, color='darkred')
plt.show()
```
