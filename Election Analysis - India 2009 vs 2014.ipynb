```python
import pandas as pd

df_09 = pd.read_csv("../Data/LS2009Candidate.csv")
df_09.head()
```

```python
df_14 = pd.read_csv("../Data/LS2014Candidate.csv")
df_14.head()
```

```python
df = pd.concat([df_09, df_14])
print(df.shape)
df.head()
```

```python
df['Party Abbreviation'].unique()
```

```python
df['Alliance'] = df['Party Abbreviation']
df['Alliance'] = df['Alliance'].replace(to_replace=['INC', 'NCP', 'RJD', 'DMK', 'IUML', 'JMM', 'JD(s)', 'KC(M)', 'RLD', 
                                                    'RSP', 'CMP(J)', 'KC(J)', 'PPI', 'MD'], value='UPA')
df['Alliance'] = df['Alliance'].replace(to_replace=['BJP', 'SS', 'LJP', 'SAD', 'RLSP', 'AD', 'PMK', 'NPP', 'AINRC', 'NPF', 
                                                    'RPI(A)', 'BPF', 'JD(U)', 'SDF', 'NDPP', 'MNF', 'RIDALOS', 'KMDK', 
                                                    'IJK', 'PNK', 'JSP', 'GJM', 'MGP', 'GFP', 'GVP', 'AJSU', 'IPFT', 'MPP', 
                                                    'KPP', 'JKPC', 'KC(T)', 'BDJS', 'AGP', 'JSS', 'PPA', 'UDP', 'HSPDP', 
                                                    'PSP', 'JRS', 'KVC', 'PNP', 'SBSP', 'KC(N)', 'PDF', 'MDPF'], 
                                        value='NDA')
df['Alliance'] = df['Alliance'].replace(to_replace=['YSRCP', 'AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 
                                                    'CPI', 'AIMIM', 'KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 
                                                    'BOPF', 'BVA', 'HJCBL', 'JVM', 'MDMK'], value='Others')
df.head()
```

```python
df.drop(columns='Party Abbreviation', inplace=True)
df.head()
```

```python
seats_win = df[df['Position'] == 1].groupby(['Alliance', 'Year'])['Position'].sum().reset_index()
seats_win = seats_win.pivot(index='Alliance', columns='Year').fillna(0)
seats_win = seats_win.sort_values([('Position', 2009), ('Position', 2014)], ascending=False)
seats_win
```

```python
import matplotlib.pyplot as plt

colors = ("orange", "green", "red", "cyan", "brown", "grey", "blue", "indigo", "beige", "yellow", "cadetblue", "khaki")

plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)

plt.pie(seats_win[('Position', 2009)], labels=seats_win.index, colors=colors, autopct='%1.1f%%')
circle = plt.Circle((0, 0), 0.7, color='white')
fig = plt.gcf()
fig.suptitle("Winning Percentages by Alliances and Major Political Parties", fontsize=14)
ax = fig.gca()
ax.add_patch(circle)

label = ax.annotate("2009", xy=(0, 0), fontsize=30, ha="center",va="center")
ax.axis('off')
ax.set_aspect('equal')
ax.autoscale_view()

plt.subplot(1, 2, 2)
plt.pie(seats_win['Position', 2014], labels=seats_win.index, colors=colors, autopct='%1.1f%%')
circle_2 = plt.Circle((0, 0), 0.7, color='white')
fig = plt.gcf()
ax = fig.gca()
ax.add_patch(circle_2)

label = ax.annotate("2014", xy=(0, 0), fontsize=30, ha="center", va="center")
ax.axis('off')
ax.set_aspect('equal')
ax.autoscale_view()

plt.show()
```

```python
cat_win = df[df['Position'] == 1].groupby(['Candidate Category', 'Year'])['Position'].sum().reset_index()
cat_win = cat_win.pivot(index='Candidate Category', columns='Year', values='Position')
cat_win
```

```python
def annot_plot(ax, w, h):
    for p in ax.patches:
        ax.annotate('{}'.format(p.get_height()), (p.get_x() + w, p.get_height() + h))
        
ax = cat_win.plot(kind='bar', title ="Winning Category", figsize=(15, 10), legend=True, fontsize=12)
annot_plot(ax, 0.05, 5)
```

```python
cat_alliance_09 = df[(df['Position'] == 1) & (df['Year'] == 2009)].groupby(['Alliance', 'Candidate Category'])['Position'].sum()
cat_alliance_09 = cat_alliance_09.unstack().fillna(0)

cat_alliance_14 = df[(df['Position'] == 1) & (df['Year'] == 2014)].groupby(['Alliance', 'Candidate Category'])['Position'].sum()
cat_alliance_14 = cat_alliance_14.unstack().fillna(0)

cat_alliance_14
```

```python
ax = cat_alliance_09.plot(kind='bar', title ="2009 Winning Category", figsize=(15, 10), legend=True, fontsize=12)
annot_plot(ax, -0.05, 1.5)

ax = cat_alliance_14.plot(kind='bar', title ="2014 Winning Category", figsize=(15, 10), legend=True, fontsize=12)
annot_plot(ax, -0.05, 1.5)
```

```python
import numpy as np

age_09 = df[(df['Position'] == 1) & (df['Year'] == 2009)]['Candidate Age']
age_14 = df[(df['Position'] == 1) & (df['Year'] == 2014)]['Candidate Age']

plt.style.use('seaborn-deep')
bins = np.linspace(20, 90, 10)

plt.hist([age_09, age_14], bins, label=['2009', '2014'])
plt.title('Distribution of Age of the winners')
plt.legend()
plt.show()
```

```python
colors = ['#0000CD', '#CD3333']

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)

plt.pie(df[(df['Position'] == 1) & (df['Year'] == 2009)]['Candidate Sex'].value_counts(), labels=['Male', 'Female'], 
        autopct='%1.1f%%', colors=colors, startangle=90)
circle = plt.Circle((0,0), 0.7, color='white')
fig = plt.gcf()
fig.suptitle("Gender Distribution in 2009 & 2014 India Elections", fontsize=14)
ax = fig.gca()
ax.add_patch(circle)
ax.annotate("2009", xy=(0, 0), fontsize=30, ha="center",va="center")
ax.axis('off')
ax.set_aspect('equal')
ax.autoscale_view()

plt.subplot(1,2,2)
plt.pie(df[(df['Position'] == 1) & (df['Year'] == 2014)]['Candidate Sex'].value_counts(), labels=['Male', 'Female'], 
        autopct='%1.1f%%', colors=colors, startangle=90)
circle_2 = plt.Circle((0,0), 0.7, color='white')
fig = plt.gcf()
ax = fig.gca()
ax.add_patch(circle_2)
ax.annotate("2014", xy=(0, 0), fontsize=30, ha="center",va="center")
ax.axis('off')
ax.set_aspect('equal')
ax.autoscale_view()

plt.show()
```

```python
elec_09 = pd.read_csv("../Data/LS2009Electors.csv")
elec_09.head()
```

```python
elec_14 = pd.read_csv("../Data/LS2014Electors.csv")
elec_14.head()
```

```python
elec_14['STATE'] = elec_14['STATE'].replace(to_replace=['Odisha'], value='Orissa')
elec_14['STATE'] = elec_14['STATE'].replace(to_replace=['Chhattisgarh'], value='Chattisgarh')
```

```python
elec_09 = elec_09.groupby('STATE')['POLL PERCENTAGE'].mean().sort_values(ascending=False).to_dict()
y_09 = [2009 for i in range(35)]
s_09 = list(elec_09.keys())
p_09 = list(elec_09.values())

elec_14 = elec_14.groupby('STATE')['POLL PERCENTAGE'].mean().sort_values(ascending=False).to_dict()
y_14 = [2014 for i in range(35)]
s_14 = list(elec_14.keys())
p_14 = list(elec_14.values())

data = {'YEAR': y_09 + y_14, 'STATE': s_09 + s_14, 'Poll_Percentage': p_09 + p_14}
df = pd.DataFrame(data=data)
df.head()
```

```python
import seaborn as sns

plt.subplots(figsize=(6, 20))
sns.barplot(x=df['Poll_Percentage'], y=df['STATE'], hue=df['YEAR'])
plt.title('Poll Percentage of States 2009 and 2014')
plt.show()
```
